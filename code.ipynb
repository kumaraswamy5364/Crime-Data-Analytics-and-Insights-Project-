```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("crime_dataset_india.csv")
```

```python
df.head(5)
```

```python
print(len(df.columns))  #checking the length and names of column
print(df.columns.tolist())
```

```python
df = df.drop(columns=['Unnamed: 14']) #drop the column names
```

```python
df.columns = [
    "Report_No", "Date_Reported", "Date_of_Occurrence", "Time_of_Occurrence",
    "City", "Crime_Code", "Crime_Description", "Victim_Age", "Victim_Gender",
    "Weapon_Used", "Crime_Domain", "Police_Deployed", "Case_Closed", "Date_Case_Closed"
] #add new names in that columns
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.describe(include='all')
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, y='City', order=df['City'].value_counts().index, palette='viridis')
plt.title("Crime Count per City")
plt.show()
```

```python
sns.countplot(data=df, x='Victim_Gender', palette='Set2')
plt.title("Gender Distribution of Victims")
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['Crime_Description'].value_counts().head(7).plot(kind='bar',color='purple')
plt.title("Top 10 Crime Types")
plt.ylabel("Count")
plt.xticks(rotation =45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Victim_Age'],bins=10, color="blue")
plt.title("Victim Age Distribution")
plt.xlabel("Age")
plt.show()
```

```python
sns.countplot(x='Case_Closed', data=df, palette='coolwarm')
plt.title("Case Closure Status")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
